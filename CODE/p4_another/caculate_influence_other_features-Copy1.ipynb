```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
influence_data = pd.read_csv('influence_data.csv')
data_by_artist = pd.read_csv('data_by_artist.csv',index_col=1,dtype={'mode':np.float64})
```

```python
data_by_artist
```

```python
data_name = data_by_artist.columns[1:13]
```

```python
influencer_acousticness = []
influencer_danceability = []
influencer_duration_ms = []
influencer_energy = []
influencer_instrumentalness = []
influencer_key = []
influencer_liveness = []
influencer_loudness = []
influencer_mode = []
influencer_speechiness = []
influencer_tempo = []
influencer_valence = []
influencer_data=[influencer_acousticness,influencer_danceability,
               influencer_duration_ms,influencer_energy,
               influencer_instrumentalness,
               influencer_key,influencer_liveness,influencer_loudness,
               influencer_mode,influencer_speechiness,
               influencer_tempo,influencer_valence]
```

```python
follower_acousticness = []
follower_danceability = []
follower_duration_ms = []
follower_energy = []
follower_instrumentalness = []
follower_key = []
follower_liveness = []
follower_loudness = []
follower_mode = []
follower_speechiness = []
follower_tempo = []
follower_valence = []
follower_data=[follower_acousticness,follower_danceability,
               follower_duration_ms,follower_energy,
               follower_instrumentalness,
               follower_key,follower_liveness,follower_loudness,
               follower_mode,follower_speechiness,
               follower_tempo,follower_valence]
```

```python

data_by_artist = np.array(data_by_artist)

for x in range(data_by_artist.shape[0]):
    if x%100==0:
        print(x/100)
    for y in range(x):
        for i in range(12):
            influencer_data[i].append(data_by_artist[x,i+1])
            follower_data[i].append(data_by_artist[y,i+1])
```

```python
result_correlation = pd.DataFrame(index=data_name,columns=['correlation']).fillna(value=0)
INDEX=range(len(follower_data[0]))
for i in range(len(data_name)):
    dinfluencer = pd.Series(influencer_data[i], index=INDEX,dtype=float)
    dfollower = pd.Series(follower_data[i], index=INDEX,dtype=float)
    df=pd.DataFrame({'influencer':dinfluencer,'followepearsonr':dfollower},dtype=float)
    result_correlation.iloc[i,0]=df.corr(method='pearson').iloc[0,1]
print(result_correlation)
```

```python
result_correlation.to_csv('pearson_correlation_of_unconnected_musicians.csv')
```

```python
for i in range(len(data_name)):
    print(i)
    dinfluencer = pd.Series(influencer_data[i], index=INDEX,dtype=float, name='influencer_'+data_name[i])
    dfollower = pd.Series(follower_data[i], index=INDEX,dtype=float, name='follower_'+data_name[i])
    df = pd.DataFrame({'influencer_'+data_name[i]:dinfluencer,'follower_'+data_name[i]:dfollower})
    df.to_csv('_balabala_'+data_name[i]+'.csv')
```
